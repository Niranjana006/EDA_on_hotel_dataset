```python
#top sales of items by quantity 3
top_item = df.groupby('Item Name')['Qty.'].sum().sort_values(ascending = False).head(3)
top_item
```

```python
#revenue wise top 3 items sold
rev_top_item = df.groupby('Item Name')['Final Total'].sum().sort_values(ascending = False).head(3)
rev_top_item
```

```python
#invoice no that spent most on 2023-09-14
sp_date = df[df['Date'] == '2023-09-14']
top_in = sp_date.groupby('Invoice No.')['Final Total'].sum().sort_values(ascending = False)
top_in.head(1)
```

```python
top_spent = df.groupby(['Date','Invoice No.'])['Final Total'].sum()
top_spent.loc['2023-09-14'].sort_values(ascending = False).idxmax()
```

```python
#return no. of invoices whose bill value is > overall avg bill value
avg_bill = df['Final Total'].sum() / df['Invoice No.'].nunique()
print(avg_bill)
In_gre_avg = df.groupby('Invoice No.')['Final Total'].sum()
print(In_gre_avg[In_gre_avg > avg_bill].count())
```

```python
# Calculate totals per invoice
invoice_totals = df.groupby('Invoice No.')['Final Total'].sum()
# Count how many totals are greater than the mean of those totals
result = (invoice_totals > invoice_totals.mean()).sum()
print(result)
```

```python
#on each date, which product was the top seller by quantity- bar
top = df.groupby(['Date','Item Name'])['Qty.'].sum().sort_values(ascending = False)
top = top.groupby('Date').head(1).reset_index()
top
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting the top seller for each date
plt.figure(figsize=(20, 6))
sns.barplot(data=top, x='Date', y='Qty.', hue='Item Name')
plt.xticks(rotation=45)
plt.title('Top Product by Quantity per Date')
plt.show()
```

```python
#do high quantity items always generate high revenue?
item_summary = df.groupby('Item Name').agg({'Qty.': 'sum', 'Final Total': 'sum'})
correlation_matrix = item_summary[['Qty.', 'Final Total']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix: Quantity vs Revenue')
plt.show()
```

```python
#which product contributes the highest % of tot rev
prod = df.groupby('Item Name')['Final Total'].sum()
tot_rev = df['Final Total'].sum()
prod_perc = prod / tot_rev * 100
prod_perc.sort_values(ascending = False).head(1)
```
